 #Ques 1
 What is the difference between a function and a method in Python?

--> A function in Python is a block of reusable code that performs a specific task and is defined using the def keyword. Functions can be called independently and are not tied to any particular object.
Example:
Function:
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))  # Output: Hello, Alice!

A method, on the other hand, is a function that is associated with an object and is called on that object. Methods are defined within a class and operate on the data contained within the object.
Example:
Method:
class Person:
    def greet(self, name):
        return f"Hello, {name}!"

p = Person()
print(p.greet("Alice"))  # Output: Hello, Alice!

# Ques 3.
 What are the different ways to define and call a function in Python?

--> Different Ways to Define and Call a Function in Python:
1. Regular Function

Define: Use the def keyword followed by the function name and parameters.
Call: Use the function name with arguments.

Example:
def add(a, b):
    return a + b

result = add(3, 5)  # Calling the function
print(result)
  # Output: 8


2. Lambda Function

Define: Use the lambda keyword for creating anonymous functions (functions without a name).
Syntax--
lambda arguments:expression

Example:
add = lambda a, b: a + b  # Defining a lambda function
result = add(3, 5)  # Calling the lambda function
print(result)  
# Output: 8


3. Function with Default Parameters

Define: Specify default values for parameters in the function definition.
Call: You can call the function with or without providing arguments for default parameters.

Example:
def greet(name="Guest"):
    return f"Hello, {name}!"

print(greet())  # Output: Hello, Guest!
print(greet("Alice"))
 # Output: Hello, Alice!


4. Function with Variable-Length Arguments

Define: Use *args for multiple positional arguments and **kwargs for multiple keyword arguments.
Call: Pass multiple arguments or keyword arguments.

Example:
def display_info(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)

display_info(1, 2, 3, name="Alice", age=25)
# Output:
# Positional arguments: (1, 2, 3)
# Keyword arguments: {'name': 'Alice', 'age':25}

#Ques 6.
 Explain the concept of generators in Python and how they are defined.

 -->
 Generators in Python are a special type of iterator that allow you to generate values one at a time using the yield keyword. Unlike regular functions, which return a single value and terminate, generators pause their execution and resume from where they left off.
How Generators Are Defined:
Generators are defined using functions with the yield keyword instead of return.

Example of a Generator for Fibonacci Sequence:
def fibonacci(n):    a, b = 0, 1    for _ in range(n):        yield a        a, b = b, a + b# Using the generatorfor num in fibonacci(5):    print(num)
Output:
0
1
1
2
3

#Ques 7.
 What are the Advantages of using Generators over Regular Functions?

-->
Memory Efficient: Generators do not store all values at once; they produce values one by one.
Faster for Large Data: Useful when working with big datasets because values are generated only when needed.
Lazy Evaluation: Values are created on demand, which saves time and resources.
Handles Infinite Sequences: Generators can work with infinite data streams.
Improves Performance: Less memory usage and no need to create large lists in advance.
Simple Iteration: Makes code cleaner and easier for loop-based processing.

#Ques 10.
 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
 -->
 map()
Work: Applies a function to each element
Output: A new iterable with modified values
Example: Multiply all numbers by 2

-filter()
Work: Selects elements that satisfy a condition (True/False)
Output: A new iterable with only filtered values
Example: Keep only even numbers

- reduce() (from functools)
Work: Combines all elements into one single value
Output: A single result
Example: Sum of all numbers

```python
#Ques 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()
```

```python
#Ques 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(1, 5), (3, 2), (4, 8), (2, 1)]

sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```

```python
#Ques 9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 10, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

```python
#Ques10.. Create a Python program that uses `filter()` to remove all the vowels from a given string
text = "Hello World"

vowels = "aeiouAEIOU"

result = "".join(filter(lambda ch: ch not in vowels, text))

print("Original:", text)
print("Without vowels:", result)
```

```python
#Ques11.) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this: Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €. Write a Python program using lambda and map.
orders = [
    [34587, 40.95, 4],
    [98762, 56.80, 5],
    [77226, 32.95, 3],
    [88112, 24.99, 10]
]

result = []

for order in orders:
    order_no, price, qty = order
    total = price * qty

    if total < 100:
        total += 10   # add 10€ extra charge

    result.append((order_no, total))

print(result)
```
